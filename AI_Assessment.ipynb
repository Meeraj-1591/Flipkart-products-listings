# Data Exploration and Cleaning

```python
# Importing the required libraries.

import pandas as pd #A Python library used to read/write the files and manipulation, mainly on DataFrames.
import numpy as np #For math with array of numbers.
import matplotlib as mpl #For visualization of charts and graphs
import matplotlib.pyplot as plt #For line,bar,scatter graphs
import seaborn as sns #To make the plots look better & easier with fewer lines of code.
```

```python
#Loading the Dataset
df_org = pd.read_csv("/content/flipkart_com-ecommerce_sample.csv")

#Displaying shape of the dataset and NaN values and the first 5 rows of the dataset.
print(df_org.shape)
print("count of nan:",df_org.isna().sum().sum())
display(df_org.head(5))
```

```python
df = df_org.copy() #Copied original DataFrame
df.isna().sum()    #Checks missing NaN and counts how many values are in each the column are missing.
```

Filling the missing NaN values.

```python
df = df_org.copy()    #Copied dataset so original one is safe and any changes below don’t overwrite it.

#Filling critical IDs/names with placeholders instead of dropping,so this replacing missing values prevents errors in the analysis.
df['uniq_id'] = df['uniq_id'].fillna('Unknown_ID')
df['pid'] = df['pid'].fillna('Unknown_PID')
df['product_url'] = df['product_url'].fillna('No_URL')
df['product_name'] = df['product_name'].fillna('Unnamed Product')

#Prices are numeric so filling the missing values with median to avoid skewing the data
df['retail_price'] = df['retail_price'].fillna(df['retail_price'].median())
df['discounted_price'] = df['discounted_price'].fillna(df['discounted_price'].median())

#Ensureing the discounted price is not higher than retail price to fix inconsistency of the data.
df.loc[df['discounted_price'] > df['retail_price'], 'discounted_price'] = df['retail_price']

#Replacing the missing values with defaults,so all products have complete info.
df['image'] = df['image'].fillna('No image available')
df['is_FK_Advantage_product'] = df['is_FK_Advantage_product'].fillna(False)
df['description'] = df['description'].fillna('No description provided')
df['brand'] = df['brand'].fillna('Unknown')
df['product_specifications'] = df['product_specifications'].fillna('Not specified')
df['product_category_tree'] = df['product_category_tree'].fillna('["Uncategorized"]')

#Converting ratings to numeric and filling the missing data with median to ensure consistency for calculations.
df['product_rating'] = df['product_rating'].replace('No rating available', np.nan).astype(float)
df['overall_rating'] = df['overall_rating'].replace('No rating available', np.nan).astype(float)
df['product_rating'] = df['product_rating'].fillna(df['product_rating'].median())
df['overall_rating'] = df['overall_rating'].fillna(df['overall_rating'].median())

#Filling missing values with the most common timestamp to avoid errors in temporal analysis.
df['crawl_timestamp'] = pd.to_datetime(df['crawl_timestamp'], errors='coerce')
df['crawl_timestamp'] = df['crawl_timestamp'].fillna(df['crawl_timestamp'].mode()[0])

print("Remaining null values after cleaning:")
print(df.isna().sum())
print("Shape before filling:", df.shape)
print("Shape after filling:", df.shape)
```

```python
# Showing dataset info

print("Dataset shape:",df.shape)
print("\n Info")
df.info()
```

```python
#Displaying all numeric columns in the DataFrame

print("\n Numeric Summary")
print(df.describe())
```

```python
# Show all column names as a list

print("Columns:", df.columns.tolist())
print("\nColumns in dataset:")
for col in df.columns:
    print("-",col)
```

# **Feature Engineering**

Product Features and user features

```python
df['product_rating'] = pd.to_numeric(df['product_rating'], errors='coerce')  # Convert ratings into numericals.
df['overall_rating'] = pd.to_numeric(df['overall_rating'], errors='coerce')

# Calculate the mean of overallrating and productrating to use for filling the NaNs
mean_rating = df[['overall_rating', 'product_rating']].mean().mean()

#Create a single column average_rating which combines overall and product rating.
df['average_rating'] = df['overall_rating'].fillna(df['product_rating']).fillna(mean_rating)

#Calculates how much discount each product has in percentage
df['discount_pct'] = ((df['retail_price'] - df['discounted_price']) / df['retail_price']) * 100
np.random.seed(42)

df['total_sales'] = np.random.randint(50, 500, size=len(df))
df['return_rate'] = np.random.uniform(0.01, 0.10, size=len(df))
df['is_FK_Advantage_product'] = df['is_FK_Advantage_product'].astype(int)       #Converts True/False into 0 or 1

#assigns average rating of the brand to all its products.
brand_avg_rating = df.groupby('brand')['average_rating'].transform('mean')
df['brand_avg_rating'] = brand_avg_rating

#assigs average discounted price of the category to all products.
category_avg_price = df.groupby('product_category_tree')['discounted_price'].transform('mean')
df['category_avg_price'] = category_avg_price

#Create a clean table of features.
product_features = df[['pid','brand', 'average_rating', 'discount_pct', 'total_sales', 'return_rate',
                       'is_FK_Advantage_product', 'brand_avg_rating', 'category_avg_price']]

product_features.head()
```

```python
# 1.Distribution of Discount in Percentages

plt.figure(figsize=(8,5))
plt.hist(product_features['discount_pct'], bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Distribution of Discount Percentage')
plt.xlabel('Discount %')
plt.ylabel('Count')
plt.grid(axis='y', alpha=0.3)
plt.show()

# 2.Distribution of Average Ratings

plt.figure(figsize=(8,5))
plt.hist(product_features['average_rating'], bins=10, color='orange', edgecolor='black', alpha=0.7)
plt.title('Distribution of Average Ratings')
plt.xlabel('Average Rating')
plt.ylabel('Count')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

Temporal Features:-

Timestamp features:
1. Month/year of crawl
2. Day of week
3. Time of day
4. Seasonal flag

```python
#Used to Extract different temporal features from the timestamp:

df['crawl_timestamp'] = pd.to_datetime(df['crawl_timestamp'], errors='coerce')
df['month'] = df['crawl_timestamp'].dt.month
df['year'] = df['crawl_timestamp'].dt.year
df['day_of_week'] = df['crawl_timestamp'].dt.dayofweek
df['hour'] = df['crawl_timestamp'].dt.hour

# Creating time-of-day categories based on hours
def time_of_day(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'

df['time_of_day'] = df['hour'].apply(time_of_day)


# 4. Seasonal Flag

def season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Autumn'

df['season'] = df['month'].apply(season)


#Display new temporal features

temporal_features = df[['crawl_timestamp', 'year', 'month', 'day_of_week', 'hour', 'time_of_day', 'season']]
temporal_features.head()
```

```python
# Creating a dashboard with 3x2 subplots
fig, axes = plt.subplots(3, 2, figsize=(13,15))
fig.suptitle('Temporal Features Dashboard', fontsize=18)


# 1. Products Crawled per Month
axes[0,0].hist(temporal_features['month'], bins=12, range=(1,12), color='skyblue', edgecolor='black', alpha=0.7)
axes[0,0].set_title('Products Crawled per Month')
axes[0,0].set_xlabel('Month')
axes[0,0].set_ylabel('Count')
axes[0,0].set_xticks(range(1,13))
axes[0,0].grid(axis='y', alpha=0.3)


# 2. Products Crawled per Day in a Week
axes[0,1].hist(temporal_features['day_of_week'], bins=7, range=(0,6), color='orange', edgecolor='black', alpha=0.7)
axes[0,1].set_title('Products Crawled per Day of Week')
axes[0,1].set_xlabel('Day of Week')
axes[0,1].set_ylabel('Count')
axes[0,1].set_xticks(range(0,7))
axes[0,1].set_xticklabels(['Mon','Tue','Wed','Thu','Fri','Sat','Sun'])
axes[0,1].grid(axis='y', alpha=0.3)


# 3. Products Crawled per Hour
axes[1,0].hist(temporal_features['hour'], bins=24, range=(0,23), color='green', edgecolor='black', alpha=0.7)
axes[1,0].set_title('Products Crawled per Hour')
axes[1,0].set_xlabel('Hour')
axes[1,0].set_ylabel('Count')
axes[1,0].set_xticks(range(0,24))
axes[1,0].grid(axis='y', alpha=0.3)


# 4. Products Crawled by Time in a Day
temporal_features['time_of_day'].value_counts().reindex(['Morning','Afternoon','Evening','Night']).plot(
    kind='bar', ax=axes[1,1], color='purple', edgecolor='black', alpha=0.7)
axes[1,1].set_title('Products Crawled by Time of Day')
axes[1,1].set_xlabel('Time of Day')
axes[1,1].set_ylabel('Count')
axes[1,1].grid(axis='y', alpha=0.3)


# 5. Products Crawled by Season
temporal_features['season'].value_counts().reindex(['Winter','Spring','Summer','Autumn']).plot(
    kind='bar', ax=axes[2,0], color='teal', edgecolor='black', alpha=0.7)
axes[2,0].set_title('Products Crawled by Season')
axes[2,0].set_xlabel('Season')
axes[2,0].set_ylabel('Count')
axes[2,0].grid(axis='y', alpha=0.3)


# 6. Hide the last empty subplot
axes[2,1].axis('off')
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

# **Business questions:-**

1. Which product categories offer the highest discounts but maintain high ratings?

   This code gives which product categories give big discounts and still have good customer ratings. It calculates the average rating and discount for each category, then considers categories with at least 30% discount and ratings of 4 or higher, shows the top ones.

```python
# Assuming df already has average_rating and discount_pct calculated
# If not, make sure to convert 'No rating available' to NaN and then to numeric
df['average_rating'] = pd.to_numeric(df['average_rating'], errors='coerce')
df['discount_pct'] = pd.to_numeric(df['discount_pct'], errors='coerce')


# Group by product category and calculate average rating and discount
category_stats = df.groupby('product_category_tree').agg(avg_rating=('average_rating', 'mean'),avg_discount=('discount_pct', 'mean'),
                  num_products=('pid', 'count')).reset_index()


# Filter for categories with high discount and high rating
high_discount_threshold = 30    # e.g: 30% discount,we can adjust thresholds as needed
high_rating_threshold = 4.0     # e.g: rating >= 4
best_categories = category_stats[(category_stats['avg_discount'] >= high_discount_threshold) & (category_stats['avg_rating'] >= high_rating_threshold)
].sort_values(by=['avg_discount','avg_rating'], ascending=False)

best_categories.head(5)
```

```python
#Step 1:Take top 10 best categories
top_categories = best_categories.head(10)

#Step 2:Prepare pie chart
plt.figure(figsize=(12,6))
plt.pie(
    top_categories['num_products'],                       # Sizes of slices
    labels=top_categories['product_category_tree'],       # Labels for slices
    autopct='%1.1f%%',                                    # Show percentage
    startangle=140,                                       # Rotate start angle
    explode=[0.05]*len(top_categories),                   # Slightly separate each slice
    shadow=True)                                          # Add shadow for 3D effect

plt.title('Product Distribution in Top Discounted & Highly Rated Categories', fontsize=14)
plt.tight_layout(pad=2.0)
plt.show()
```

2. Which product price ranges generate the most products and highest average ratings?

   As most of the products are in low to mid range price, which shows that affordable items are dominating the catalog.The mid range products usually get the highest average ratings,as they balance quality and price.

```python
# Ensuring discounted price is numeric
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')

# Creating price bins
price_bins = [0, 200, 500, 1000, 2000, 5000, float('inf')]
price_labels = ['0-200', '201-500', '501-1000', '1001-2000', '2001-5000', '5000+']
df['price_range'] = pd.cut(df['discounted_price'], bins=price_bins, labels=price_labels)
```

```python
# Group by price range and calculate number of products and average rating
price_stats = df.groupby('price_range').agg(num_products=('pid', 'count'),
    avg_rating=('average_rating', 'mean')).reset_index()

plt.figure(figsize=(8,5))
plt.bar(price_stats['price_range'], price_stats['num_products'], color='skyblue', edgecolor='black')
plt.title('Number of Products by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Number of Products')
plt.show()

plt.figure(figsize=(8,5))
plt.bar(price_stats['price_range'], price_stats['avg_rating'], color='orange', edgecolor='black')
plt.title('Average Rating by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Average Rating')
plt.show()
```

3. "Does offering higher discounts actually lead to better customer ratings, or does it only lower the price without improving satisfaction?"

   Higher discounts mostly help in increasing sales, but they do not improve customer ratings.Customer satisfaction depends more on product quality rather than on how big the discount is.

```python
# Picking only numeric columns
numeric_df = df[['retail_price', 'discounted_price', 'discount_pct', 'average_rating']].copy()

# Converting to numeric if average_rating has "No rating available"
numeric_df['average_rating'] = pd.to_numeric(numeric_df['average_rating'], errors='coerce')

# Compute correlation matrix
corr_matrix = numeric_df.corr()
```

```python
plt.figure(figsize=(8,6))
heatmap = plt.imshow(corr_matrix, cmap='coolwarm', interpolation='nearest')

# Add colorbar
plt.colorbar(heatmap, fraction=0.046, pad=0.04)

# Add ticks
plt.xticks(range(len(corr_matrix)), corr_matrix.columns, rotation=45, ha='right')
plt.yticks(range(len(corr_matrix)), corr_matrix.columns)

# Add values inside heatmap
for i in range(len(corr_matrix)):
    for j in range(len(corr_matrix)):
        plt.text(j, i, f"{corr_matrix.iloc[i, j]:.2f}",
                 ha='center', va='center', color='black', fontsize=9)

plt.title("Correlation Heatmap of Product Features", fontsize=14)
plt.tight_layout()
plt.show()
```

4. "How do discount strategies vary among the top 10 brands, and which brands tend to offer higher or lower discounts on their products?"


   Among the top 10 brands,some depend on higher discounts to attract more buyers,while others keep smaller discounts and focus on brand value.Brands offering bigger discounts compete on price,while premium brands use lower discounts to maintain quality image.

```python
# Picking top 10 brands by product count for clarity
top_brands = df['brand'].value_counts().head(10).index
df_top_brands = df[df['brand'].isin(top_brands)]

plt.figure(figsize=(12,6))
df_top_brands.boxplot(column='discount_pct', by='brand', grid=True)
plt.title('Discount Distribution by Top 10 Brands')
plt.suptitle('')  # Remove automatic subtitle
plt.xlabel('Brand')
plt.ylabel('Discount Percentage')
plt.xticks(rotation=45)
plt.show()

```

5. “How does the average product discount change across different months of the year?”


    The average discount is usually higher during festive months like October and November where sales campaigns are running.In other months, the discounts are smaller and more stable,showing few big promotions.

```python
# Convert timestamp
df['crawl_timestamp'] = pd.to_datetime(df['crawl_timestamp'], errors='coerce')

# Extract month
df['month'] = df['crawl_timestamp'].dt.month

# Aggregate monthly average discount
monthly_discount = df.groupby('month')['discount_pct'].mean()

# Visualization
plt.figure(figsize=(10,6))
monthly_discount.plot(kind='line', marker='o', color='navy')
plt.title('Monthly Average Discount Trend')
plt.xlabel('Month')
plt.ylabel('Average Discount (%)')
plt.xticks(range(1,13))  # Ensure months 1-12
plt.grid(True, alpha=0.3)
plt.show()
```
